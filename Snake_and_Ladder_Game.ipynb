```python
import random
```

```python
def dice():
    return random.randint(1,6)
```

```python
def player1(pos=0):
    while pos<1:
        input("Player 1: Press enter to throw the dice: ")
        throw=dice()
        print(throw)
        if throw==6:
            pos=1
            return pos
        else:
            return pos
    while pos>0 and pos<=100:
        input("Player 1: Press enter to throw the dice: ")
        throw=dice()
        print(throw)    
        pos+=throw
        if pos>100:
            pos-=throw
            print("Throw out of range. Better luck next time")
        pos=check_position(pos)
        return pos
```

```python
def player2(pos=0):
    while pos<1:
        input("Player 2: Press enter to throw the dice: ")
        throw=dice()
        print(throw)
        if throw==6:
            pos=1
            return pos
        else:
            return pos
    while pos>0 and pos<=100:
        input("Player 2: Press enter to throw the dice: ")
        throw=dice()
        print(throw)
        pos+=throw
        if pos>100:
            pos-=throw
            print("Throw out of range. Better luck next time")
        pos=check_position(pos)
        return pos
```

```python
def check_position(pos=0):
    snake={33:5,54:34,63:16,93:74,97:61}
    ladder={2:38,8:31,21:42,46:84,51:67,71:91,80:99}
    for x in snake:
        if pos==x:
            print("Oops! A snake. You are going back to",snake[x])
            pos=snake[x]
            break
        else:
            continue
    for y in ladder:
        if pos==y:
            print("Yay!! A ladder. You are moving up to",ladder[y])
            pos=ladder[y]
            break
        else:
            continue
    return pos
```

```python
x=input("Press s to start the game: ")
a=0;b=0
while x=="s" or x=="S":
    
    p1=player1(a)
    a=p1
    print(p1)
    p2=player2(b)
    b=p2
    print(p2)
    if p1>=100:
        print("Congratulations player 1 you have won the match!!")
        x="e"
        break
    elif p2>=100:
        print("Congratulations player 2 you have won the match!!")
        x=="e"
        break
    else:
        continue
```
